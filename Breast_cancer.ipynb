```python
import pandas as pd
import numpy as np
import sklearn.datasets
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

Data collection and preprocessing

```python
#loading the data from sklearn
data=sklearn.datasets.load_breast_cancer()
print(data)
breast_data=pd.DataFrame(data.data,columns=data.feature_names)
#print(data.head())
breast_data['label']=data.target
breast_data.tail()
breast_data.isnull().sum()
breast_data.describe()
breast_data.info()
```

```python
breast_data['label'].value_counts()
breast_data.groupby('label').mean()
x=breast_data.drop(columns='label',axis=1)
y=breast_data['label']
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=2,stratify=y,test_size=0.2)
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
pred1=model.predict(x_train)
pred2=model.predict(x_test)
print("accuracy",accuracy_score(y_train,pred1))
print("accuracy",accuracy_score(y_test,pred2))
```
